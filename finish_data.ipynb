```python
df = pd.read_csv("./failure_predictions_1_copy.csv")

def calculate_risk(row):
    # Engine Temperature
    if row['Component'] == "Engine" and row['Parameter'] == "Temperature":
        if row['Value'] < 105:
            return "No"
        else:
            return "Yes"
        
    # Engine Oil Pressure
    elif row['Component'] == "Engine" and row['Parameter'] == "Oil Pressure":
        if 25 < row['Value'] < 65:
            return "No"
        else:
            return "Yes"
        
    # Engine Speed
    elif row['Component'] == "Engine" and row['Parameter'] == "Speed":
        if row['Value'] < 1800:
            return "No"
        else:
            return "Yes"
        
    # Drive, Brake Control
    elif row['Component'] == "Drive" and row['Parameter'] == "Brake Control":
        if row['Value'] > 1:
            return "No"
        else:
            return "Yes"
        
    # Drive, Transmission Pressure
    elif row['Component'] == "Drive" and row['Parameter'] == "Transmission Pressure":
        if 200 < row['Value'] < 450:
            return "No"
        else:
            return "Yes"
    
    # Drive, Pedal Sensor
    elif row['Component'] == "Drive" and row['Parameter'] == "Pedal Sensor":
        if row['Value'] < 4.7:
            return "No"
        else:
            return "Yes"
    
    # Fuel, Water in Fuel
    elif row['Component'] == "Fuel" and row['Parameter'] == "Water in Fuel":
        if row['Value'] < 1800:
            return "No"
        else:
            return "Yes"
    
    # Fuel, Level
    elif row['Component'] == "Fuel" and row['Parameter'] == "Level":
        if row['Value'] > 1:
            return "No"
        else:
            return "Yes"
        
    # Fuel, Pressure
    elif row['Component'] == "Fuel" and row['Parameter'] == "Pressure":
        if 35 < row['Value'] < 65:
            return "No"
        else:
            return "Yes"
        
    # Fuel, Temperature
    elif row['Component'] == "Fuel" and row['Parameter'] == "Temperature":
        if row['Value'] < 400:
            return "No"
        else:
            return "Yes"
        
    # Misc, System Voltage
    elif row['Component'] == "Misc" and row['Parameter'] == "System Voltage":
        if 12 < row['Value'] < 15:
            return "No"
        else:
            return "Yes"
        
    # Misc, Exhaust Gas Temperature
    elif row['Component'] == "Misc" and row['Parameter'] == "Exhaust Gas Temperature":
        if row['Value'] < 365:
            return "No"
        else:
            return "Yes"
        
    # Misc, Hydraulic Pump Rate
    elif row['Component'] == "Misc" and row['Parameter'] == "Hydraulic Pump Rate":
        if row['Value'] < 125:
            return "No"
        else:
            return "Yes"
        
    # Misc, Air Filter Pressure
    elif row['Component'] == "Misc" and row['Parameter'] == "Air Filter Pressure":
        if row['Value'] > 20:
            return "No"
        else:
            return "Yes"
    # Fits into none of the categories   
    else:
        return -1   
    

df['Risk'] = df.apply(calculate_risk, axis=1)
```

```python
count_1 = 0
count_0 = 0
count_11 = 0
for elem in df['Risk']:
    if elem == 1:
        count_1 += 1
    elif elem == -1:
        count_11 += 1
    else:
        count_0 += 1
        
print(f"count1: {count_1}, count0: {count_0}, count11: {count_11}")
```

```python
def get_normal_time(iso_time):
    from datetime import datetime
    datetime_obj = datetime.strptime(iso_time, "%Y-%m-%dT%H:%M:%SZ")
    human_readable = datetime_obj.strftime("%B %d, %Y %I:%M:%S %p UTC")
    return human_readable
```

```python
for i, elem in enumerate(df['Time']):
    df.loc[i, 'Time'] = get_normal_time(elem)
```
